```python
# import library
import requests 
from bs4 import BeautifulSoup
```

```python
results = requests.get('https://warstek.com/badak-putih-utara/') 
results.text 
```

```python
# Encoding
results.encoding = results.apparent_encoding 
results.encoding
```

```python
# Parsing 
WP = BeautifulSoup(results.text,'html.parser') 
```

```python
WP.find('p') 
```

```python
WP.find('p').getText() 
```

```python
WP.find_all('p') 
```

```python
#Buat varibael untuk menyimpan data
Full_Text = [] 
for paragraf in WP.find_all('p'): 
    Full_Text.append(paragraf.getText()) 

Full_Text 
```

```python
#Membuang White Space
Full_Text = [paragraf.strip() for paragraf in Full_Text if paragraf !=''] 
Full_Text 
```

```python
# Menggabungkan Element
Full_Text = ' '.join(Full_Text) 
Full_Text 
```

# Preprocessing Text Bahasa Indonesia

```python
text = Full_Text 
```

```python
# Lower Text
text = text.lower() 
text 
```

```python
import re 
```

```python
# Menghilangkan Simbol atau Karakter
text = re.sub(r'[-()"/\-@;:<>{}`+=~|.!?,]','', text) 
text 
```

```python
import nltk 
from nltk import word_tokenize
nltk.download('punkt') 
nltk.download('stopwords') 
nltk.download('wordnet') 
```

```python
# Tokenizing 
token = word_tokenize(text) 
print(token)
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory 
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory 
```

```python
# Hapus StopWord
stopword_remover = StopWordRemoverFactory() 
stemmer_remover = StemmerFactory() 

ina_stemmer = stemmer_remover.create_stemmer() 
ina_stopword = stopword_remover.create_stop_word_remover() 

stopword_remover.get_stop_words() 
```

```python
# Membuat Kalimat Baru dari Token yang sudah dibuat sebagai parameter yang digunakan pada fungsi ina_stopwordrmove
text_from_token = ' '.join(token) 
text_from_token 

text_from_token_nostopwords = ina_stopword.remove(text_from_token) 
```

```python
text_from_token_nostopwords 
```

```python
# Stemming 
stemmed_text = ina_stemmer.stem(text_from_token_nostopwords) 
stemmed_text 
```

```python
# Tokenize 
token_from_stem_text = word_tokenize(stemmed_text) 
token_from_stem_text 
```

# BAG OF WORDS Bahasa Indonesia

```python
frekuensi_kata_ina = {} 
for kata in token_from_stem_text: 
    if kata not in frekuensi_kata_ina: 
        frekuensi_kata_ina[kata]=1 
    else: 
        frekuensi_kata_ina[kata]+=1 

frekuensi_kata_ina 
```
